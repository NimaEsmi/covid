```python
from pathlib import Path
import shutil

file_path = Path('/Users/yasaman/Desktop/Dataset/archive_2/train.txt')
negative_file_names = []
positive_file_names = []

with open(file_path) as f:
    c_n = 0
    c_p = 0
    for line in f:
        if 'negative' in line:
            if c_n < 750:
                Dest_dir = Path('/Users/yasaman/Desktop/Dataset') / 'Negetives'
                negative_file_names.append(line.split()[1])
                c_n += 1
                
        elif 'positive' in line:
            if c_p < 750:
                Dest_dir = Path('/Users/yasaman/Desktop/Dataset') / 'Positives'
                positive_file_names.append(line.split()[1])
                c_p += 1
        
        if c_n >= 750 and c_p >= 750:
            break
            
        Dest_dir.mkdir(exist_ok=True)
```

```python
for i in range(len(negative_file_names)):
    image_path = train_path / negative_file_names[i]
    if image_path.exists():
       shutil.copy(str(image_path), '/Users/yasaman/Desktop/Dataset/Negetives')
```

```python
for i in range(len(positive_file_names)):
    image_path = train_path / positive_file_names[i]
    if image_path.exists():
       shutil.copy(str(image_path), '/Users/yasaman/Desktop/Dataset/Positives')
```

```python
import numpy as np
from pathlib import Path
from skimage import io, img_as_float # 17, 18
import imquality.brisque as Brisque
import shutil

dest_dir = Path('/Users/yasaman/Desktop/Dataset/Negetives')

n = 0
for item in dest_dir.iterdir():
    check_path = Path(f'/Users/yasaman/Desktop/Dataset/Negatives/{item.name}')
    n += 1
    if check_path.exists():
        continue
        
    if not check_path.exists():
        img=img_as_float(io.imread(item, as_gray=True)).astype(np.float32)
        score=Brisque.score(img)
    
        if score < 40:
           shutil.copy(str(item), '/Users/yasaman/Desktop/Dataset/Negatives')
        
    if n % 10 == 0:
        print('is loading ...')
```
